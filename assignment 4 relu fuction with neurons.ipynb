```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def relu(x):
    return np.maximum(0,x)

X = np.arange(-10,10,0.1)
```

```python
f1 = X**3 + X**2 + X + 1
plt.plot(X,f1)
plt.grid()
```

```python
f2 = relu((-5 * X ) - 7.7)
plt.plot(X,f2)
plt.title("Neuron 1")
plt.grid()
```

```python
f3 = relu((-1.2 * X) - 1.3)
plt.plot(X,f3)
plt.title("Neuron 2")
plt.grid()
```

```python
f4 = relu((1.2 * X) + 1)
plt.plot(X,f4)
plt.title("Neuron 3")
plt.grid()
```

```python
f5 = relu((1.2 * X) - 0.2)
plt.plot(X,f5)
plt.title("Neuron 4")
plt.grid()
```
